```python
#img rotation
from PIL import Image
import numpy as np
from scipy import ndimage
from matplotlib import pyplot as plt
%matplotlib inline
a=Image.open('img.jpg')
b=np.array(a)
c=ndimage.rotate(b,45,cval=110)
d=Image.fromarray(c)
d.save('img.jpg')
plt.imshow(d)
```

```python
#image scaling
from PIL import Image
import numpy as np
from scipy import ndimage
from matplotlib import pyplot as plt
%matplotlib inline
a=Image.open('img.jpg')
b=np.array(a)
c=ndimage.zoom(b,(.5,1,1))
d=Image.fromarray(c)
d.save('img.jpg')
plt.imshow(d)
```
