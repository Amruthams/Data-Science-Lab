```python
import pandas as pd
a=pd.read_csv('vt.csv')
print(a.info())
print(a)

#empty cells
mean=a['Calories'].mean()
a['Calories'].fillna(mean,inplace=True)
print("\n\nEmpty cells\n",a)

#date
a['Date']=pd.to_datetime(a['Date'])
print("\n\nDate\n",a)

#datatype
print("\n\nDatatype\n")
a['Maxpulse']=a['Maxpulse'].astype('int64')
print("\n\n",a)

a['Calories']=a['Calories'].astype('int64')
print("\n\n",a)

#Duration
a.loc[3,'Duration']=45
print("\n\nDuration\n",a)

#Duplication
b=a.duplicated().any()
print("\n\nDuplicate datas are found\n",b)
c=a.drop_duplicates()
print("\n\nDuplicate datas are removed\n",c)
```
